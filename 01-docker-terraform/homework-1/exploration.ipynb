```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')

parquetname = "green_tripdata_2025-11.parquet"
csvname = "taxi_zone_lookup.csv"
```

```python
df_trip = pd.read_parquet(parquetname)
print(df_trip.head())
```

```python
print(df_trip.dtypes)
```

```python
df_zone = pd.read_csv(csvname)
print(df_zone.head())
```

```python
print(df_zone.dtypes)
```

```python
print(pd.io.sql.get_schema(df_trip, name='green_taxi_data', con=engine))
```

```python
print(pd.io.sql.get_schema(df_zone, name='taxi_zone', con=engine))
```

```python
print(df_trip.to_sql(name='green_taxi_trip', con=engine, if_exists='replace'))
print(df_zone.to_sql(name='taxi_zone', con=engine, if_exists='replace'))
```
